```python
#merging  dataframe Telcom,telcom2 and 3
Telcom123 = pd.concat((Telcom, Telcom2,Telcom3))
Telcom123.head()
```

```python
TELCOM=Telcom123.drop(['DW_A_NUMBER_INT', 'DW_B_NUMBER_INT',"COUNTRY_A","COUNTRY_B","CELL_ID","CELLID","SIET_ID","DW_B_NUMBER","DW_A_NUMBER"], axis = 1) 
TELCOM.head(5)
```

```python
TELCOMCELL=CELL.merge(TELCOM,how="left", on="SITE_ID")
TELCOMCELL.head(5)
```

#Data analysis


#Which ones were the most used city for the three days?

```python
count = TELCOMCELL['VILLES'].value_counts() 
count.head(5)
```

#Which cities were the most used during business and home hours?

```python
CITY=TELCOMCELL["VILLES"].groupby(TELCOMCELL["DATETIME"])
CITY
citymax=CITY.max().head(10)
citymax
count = citymax.value_counts() 
count.head(5)
```
